How many apps are paid?

```python
paid_apps = df_new_features[df_new_features['free'] == False]
```

```python
len(paid_apps)
```

```python
Number_of_paid_apps =  len(paid_apps)

Number_of_free_apps = len(df) - Number_of_paid_apps
```

```python
print("Number of paid apps : ",len(paid_apps))
```

```python
print("Number of Free apps : ",Number_of_free_apps)
```

We can conclude that out of sample size 250, total 6 apps are paid which is 2.4% only

```python
from matplotlib import pyplot as plt

# Sample data
Number_of_paid_apps = len(paid_apps)
Number_of_free_apps = len(df) - Number_of_paid_apps

# Create bar chart
plt.figure(figsize=(5, 6))
bars = plt.bar(('Paid_apps', 'Free_apps'), [Number_of_paid_apps, Number_of_free_apps], alpha=0.5)

# Add counts on top of each bar
for bar, count in zip(bars, [Number_of_paid_apps, Number_of_free_apps]):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height() + 0.05,
             f'{count}',
             ha='center',
             va='bottom')

plt.ylabel('Number of apps')
plt.title('Paid vs Free apps in Playstore')

plt.show()
```

```python
paid_apps['released_year'] = paid_apps['released'].apply(lambda x: x.year)
```

```python
paid_apps['released_year'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Assuming 'released_year' is the column name in your DataFrame
plt.figure(figsize=(8, 6))
plt.hist(paid_apps['released_year'], bins=10, color='green', edgecolor='black')
plt.xlabel('Released Year')
plt.ylabel('Frequency')
plt.title('Distribution of Released Years in Paid Apps')
plt.grid(True)
plt.show()
```
